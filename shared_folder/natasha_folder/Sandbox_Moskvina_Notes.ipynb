Мужчины vs Женщины:

• Распределение жира:  
  М: абдоминальное (живот) ← опаснее для здоровья!  
  Ж: гиноидное (бедра, ягодицы).  

• Мышечная масса:  
  М: в среднем +40% мышц.  
  Ж: в среднем +10% жира при том же ИМТ.  

• Метаболизм:  
  М: базовый метаболизм выше на 5-10%. 
  Ж: гормональные колебания (цикл, менопауза).
